```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
tv_hours = [3,5,2,0,5,3,1,4,3,4]  # öğrencilerin TV izleme süreleri
GPA = [2.7,2.1,3.3,3.4,2.0,3.0,3.6,2.8,3.5,2.6]  # not ortalaması
```

```python
# Scatter plot = iki numeric veri arasındaki görsel ilişkiyi gösterir.
```

```python
plt.scatter(tv_hours, GPA);
```

```python
plt.figure(figsize=(8,6))
ax = plt.axes()
ax.scatter(tv_hours, GPA)
ax.set_xlabel("TV Hours")
ax.set_ylabel("GPA")
ax.axis("tight")
plt.show()
```

```python
# convert the lists to numpy arrays
```

```python
x =np.array(tv_hours)
y=np.array(GPA)
```

```python
# Use Pearson Correlation Formula
```

```python
# listeler arasında işlem yapabilmek için array'a çevirdim
len(x)*sum(x*y) - sum(x)*sum(y)
```

```python
np.sqrt((len(x)*sum(x**2)-sum(x)**2)*(len(y)*sum(y**2)-sum(y)**2)) # payda kısmı
```

```python
# pay/payda = pearson korelasyon katsayısı
-71/82.84926
```

```python
# scipy stats modu
r, p =stats.pearsonr(tv_hours, GPA)
#ilk değeri r dir, ikincisi p: p value
```

```python
r
```

```python
p #korelasyonun anlamlılığını ifade ediyor
```

The Spearman rank-order correlation coefficient is a nonparametric measure of the monotonicity of the relationship between two datasets.

Unlike the Pearson correlation, the Spearman correlation does not assume that both datasets are normally distributed.



```python
#spearman non parametric bir ölçümdür, parametriklerde assumptionlar vardır, gerçek değerler vardır.
```

```python
stats.spearmanr(tv_hours, GPA) #çok kullanılmaz
```

Calculate Kendall's tau, a correlation measure for ordinal data.
Kendall's tau is a measure of the correspondence between two rankings.

```python
stats.kendalltau(tv_hours, GPA)
```

# Numpy

```python
#calculate covariance with Numpy
np.cov(GPA, tv_hours)
#-0.78 x ve y değişkenlerinin arasındaki kovaryansını veriyor. kovaryans'ın sadece işareti işimize yarıyor
# 0.31 GPA'in varyansı, 2.66 tv_hours'un varyansı
```

```python
stats.tvar(GPA) #GPA's SAMPLE variance
```

```python
np.var(GPA) #GPA's population variance
```

```python
stats.tvar(tv_hours)
```

calculate the correlation coefficients with Numpy

```python
np.corrcoef(tv_hours, GPA) # 1'ler kendisi ile olan corelation
# return Pearson correlation
```
